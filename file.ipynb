```python
# String formatting
s = ""
n = int(input("Enter n = "))
file = open('Sangam1.txt','a')
for i in range(n):
    name = input("Enter name = ")
    phone = int(input("Enter phone = "))
    info = f"{name} {phone}\n"
    s = s+info
    
print(s)
file.write(s)
file.close()
```

```python
# String formatting
s = ""
n = int(input("Enter n = "))
file = open('Sangam.csv','w')
for i in range(n):
    name = input("Enter name = ")
    phone = int(input("Enter phone = "))
    info = f"{name},{phone}\n"
    s = s+info
    
print(s)
file.write('Name,Phone\n')
file.write(s)
file.close()
```
