```python
x = int(input("Enter x = "))
for i in range(1,11):
    print(x*i)
```

```python
x = int(input("Enter x = "))
for i in range(1,11):
    print(x,"*",i,"=",x*i)
```

```python
#  sum of n numbers
s = 0
n = int(input("Enter n = "))
for i in range(n):
    x = int(input("Enter x = "))
    s = s+x
    print(s)
```

```python
# sum of n numbers
s = ""
n = int(input("Enter n = "))
for i in range(n):
    x = input("\nEnter x = ")
    s = s+x+""
    print(s)
```

```python
# sum of n value
s = ""
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    phone = input("Enter phone = ")
    s = s+name+" "+phone+"\n"
print(s)
```
