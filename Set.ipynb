```python
# set
# Not Indexed
# No Duplicate Values
# Unordered
# Mutable
```

```python
# set are Unordered
s =  set()
a = {"Apple","Ball",'Cat',"Dog"}
print(a)
```

```python
# Set are not Indexed
a = {"Apple","Ball",'Cat',"Dog"}
print(a[0])
```

```python
# Set donot take duplicate values

a = {"Apple","Ball","Cat","Dog","Fish","Apple"}
print(a)
```

```python
# Set is Mutable
a = {"Apple","Ball",'Cat',"Dog"}
a.update({'Goat'})
a
```

```python
# Mutable
a = {"Apple","Ball",'Cat',"Dog"}
a.add('Goat')
a
```

```python
# Set is Unordered
s = set()
n = int(input("Enter n = "))
for i in range(n):
    x = input("Enter x = ")
    s.add(x)
print(s)
```

```python
# Union
abc  = {"Ram","Shyam","Hari","Gita"}
xyz = {"Siva","Sita","Ram","Shyam"}
abc.union(xyz)
```

```python
# Intersection
abc  = {"Ram","Shyam","Hari","Gita"}
xyz = {"Siva","Sita","Ram","Shyam"}
abc.intersection(xyz)
```

```python
# Associated with abc compnay but not xyz
abc  = {"Ram","Shyam","Hari","Gita"}
xyz = {"Siva","Sita","Ram","Shyam"}
abc.difference(xyz)
```

```python
# Assocaited with xyz company but not abc
abc  = {"Ram","Shyam","Hari","Gita"}
xyz = {"Siva","Sita","Ram","Shyam"}
xyz.difference(abc)
```
