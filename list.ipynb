```python
# # Matrix
# a = [[1,2,3],
#     [4,5,6],
#     [7,8,9]]
# r = ?
# c = ?
[[1,2,3],[4,5,6]]
```

```python
# WAP to develop a matrix (CRUD)

a = []
c  = int(input("Enter c = "))
r = int(input("Enter r = "))
for i in range(r):
    b  = []
    for j in range(c):
        x = int(input("Enter x = "))
        b.append(x)
    a.append(b)
    
print(a)
```

```python
# WAP to change 
z = "Apple 200 3 600\nBanana 100 2 200"
y = z.split('\n')
print(y)
# ['Apple 200 3 600', 'Banana 100 2 200']
#to
# [['Apple',200,3,600],['Banana','100,2,100']]
b = []
for i in y:
    x = i.split()
    b.append(x)
    
print(b)
```
