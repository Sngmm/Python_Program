```python
# # tuple inside tuple
# a = ((1,2,3),
#      (4,5,6),
#      (7,8,9))
```

```python
data = ()
n = int(input("Enter n = "))
for i in range(1,n+1):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter add = ")
    info = (i,name,age,add)
    data = data+(info,)
    
print(data)
```

```python
# # tuple inside tuple to list inside list
# list inside tuple
# tuple inside tuple to list inside list

a = ((1,2,3)) , ((4,5,6)) ,((7,8,9))
b = [list(x) for x in a]
print(b)
```
