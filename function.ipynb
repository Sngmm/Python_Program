```python
l = int(input("Enter l = ")) #global variable
b = int(input("Enter b = "))
h = int(input("Enter h = "))

def area():
    a = l*b
    print("The area is = ",a)
    
def volume():
    v = l*b*h
    print("The volume is = ",v)
    
volume()
area()
# area()
```

```python
# def <function_name>(<parameter>):
#     operations
#     ...
#     ...
# <function_name>(<argument>)
```

```python
def hello(x):
    print(x)
hello("Hello World")
```

```python
def cal(x,y,z):
    a = x*y
    v = a*z
    print(a,v)
cal(200,34,2)
```

```python
# function with  argument
def cal(x,y,z):
    a = x*y
    v = a*z
    print("The area is = ",a)
    print("The volume is = ",v)
l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
cal(l,b,h)
```

```python
# Return type function

def hello(x):
    return x
y = hello("Hello World")
print(y)
```

```python
def area(x,y):
    a = x*y
    return a
ar = area(10,2)
print("The area is",ar)
```

```python
def cal(x,y,z):
    a = x*y
    v = a*z
    return a,v
ar = cal(10,2,2)
area,volume = ar
print("The volume is = ",volume)
print("The area is = ",area)
```
